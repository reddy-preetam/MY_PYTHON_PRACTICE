**High-level data types are advanced data types used to store multiple values.

They are built-in and easy to use.

They can store different types of data together.

These data types make programs more powerful and flexible.

List – stores ordered and changeable items.

Tuple – stores ordered but unchangeable items.

Set – stores unordered and unique items only.

Dictionary – stores data in key–value pairs.

They support looping, indexing, and methods.

Used in real-world applications like data processing and d**

**2______**

**List is a collection of multiple items.

It is ordered (fixed position).

It is changeable (mutable).

Written using square brackets [ ].

Allows duplicate values.

Supports indexing and slicing.

Can store different data types.

Used when data needs frequent changes.

Common methods: append(), remove**

**3______**

**Tuple is a collection of items.

It is ordered.

It is unchangeable (immutable).

Written using round brackets ( ).

Allows duplicate values.

Faster than list.

Supports indexing.

Used for fixed data.

Safer than lis

**4_______**

**"""Dictionary stores data in key–value pairs.

Written using curly braces { }.

Keys must be unique.

Values can be duplicate.

It is changeable.

Data is accessed using keys.

Very fast for searching.

Used in real-world data storage.

Example methods: keys(), values()**

**5______**

**Set is a collection of unique elements.

It is unordered.

Does not allow duplicates.

Written using curly braces { }.

It is changeable.

No indexing allowed.

Used for removing duplicates.

Supports set operations.

Faster for membership check**

```python
def my_range(n):
    i=0
    while i<n:
        yield i
        i+=2
t=tuple(my_range(20))
print(t)
```

```python
# 🔹 Union (|)

# 👉 A aur B dono ke saare elements
# 👉 Common + alag-alag sab include
```

```python
# 🔹 Intersection (&)

# 👉 Sirf common elements

# Matalab:

# Jo A aur B dono me
```

```python
# 🔹 Symmetric Difference (^)

# 👉 Jo common nahi ho

# Matalab:

# A aur B ke elements, lekin common hata
```

```python
# 🔹 Difference (-)

# 👉 A me ho, par B me na ho

# Matalab:

# A ke unique elements

# ( A - B ≠ B - A )
```
