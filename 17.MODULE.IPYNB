 MODULE in Python 

1️ Module ek Python file (.py) hoti hai

2️ Module me functions, variables, classes hote hain

3️ Module ka use code reuse ke liye hota hai

4️ Module program ko organized banata hai

5️Module ko import keyword se use karte hain

6️ Built-in modules jaise math, random, os hote hain

7️ User-defined module khud create kar sakte hain

8️ Module ka naam file ke naam jaisa hota hai

9️ as keyword se module ko alias de sakte hain

Module large programs ko easy to manage banata hai

```python
# math_module.py
def add(a, b):
    return a + b
```

```python
import math
print(math.sqrt(16))
```

```python
from math import sqrt
print(sqrt(25))
```

```python
from math import sqrt, pow
print(pow(2, 3))
```

```python
import math as m # ALIAS KA MATALAB VAR SAMLLER NAME DENA
print(m.pi)
```

```python
def greet():
    print("Hello User")
```

```python
import random
print(random.randint(1, 10))
```

They add extra features (data science, web, AI, etc.).

Widely used in real-world projects.

Maintained by open-source community.

Can be updated or removed easily.

Work across platforms (Windows, Linux, Mac).

Huge collection available on PyPI.
