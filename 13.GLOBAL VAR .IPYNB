A global variable is a variable declared outside any function.

It can be accessed anywhere in the program.

Global variables have global scope.

They are created when the program starts execution.

They exist until the program ends.

A function can read a global variable without any keyword.

To modify a global variable inside a function, the global keyword is required.

Using many global variables can make code hard to debug.

Local variables have higher priority inside a function.

Global variables are useful for shared data across functions.

```python
x=10

def call():
    print(x)
call()
print(x)
```

```python
x=10
def change():
    global x
    x=20
change()
print(x)
```

 What is Global Shadowing?

 Global shadowing happens when a local variable inside a function has the same name as a global variable.

The local variable hides (shadows) the global variable inside the function.

Outside the function, the global variable remains unchanged

Global Shadowing

Shadowing occurs when local and global variables have the same name.

Inside the function, the local variable is used first.

The global variable is not modified automatically.

Shadowing affects only the function scope.

After the function call, the global variable keeps its original value

```python
x = 100   # global variable

def demo():
    x = 50   # local variable (shadows global x)
    print("Inside function:", x)

demo()
print("Outside function:", x)
```

```python
x = 100

def demo():
    global x
    x = 50
    print("Inside function:", x)

demo()
print("Outside function:", x)
```
