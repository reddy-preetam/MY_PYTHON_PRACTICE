**A while loop is a loop that repeatedly executes a block of code as long as a given condition is True.

 When the condition becomes False, the loop stops automatically
 A while loop is used to repeat code.

 It runs as long as a condition is True.

 The condition is checked before each loop run.

 If the condition becomes False, the loop stops.

It is useful when the number of repetitions is not fixed.

 It needs an update (increment/decrement) to avoid infinite loop
 
repeatedly excute block of code as long as the specified condition is true **

```python
count=0
while count < 10:
    count+=1
    print(count)
```

```python
count=0
while True:  #INFINITE LOOP
    print(count)
    count+=1
    if count >10:
        break
```

```python
is_running=True
while is_running:
    user_input=input("enter 'stop' to exit")
    if user_input=='stop':
        is_running=False
        print("LOOP DONE")
```

```python
import time

start_time=time.time()  # IT MEANS SECONDS 

timeout_second =2

while time.time()-start_time <timeout_second:
    print(" LOOP IS RUNING")
```
