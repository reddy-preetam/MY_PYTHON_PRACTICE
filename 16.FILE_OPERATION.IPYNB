What are File Operations?

File operations are used to create, open, read, write, and close files stored on disk.

 Why File Operations?

Store data permanently

Read data later

Handle large data efficiently

 FILE OPERATIONS STEPS

1️Open a File

2️Read / Write / Append

3️ open() Function

file = open("data.txt", "r")

 File Modes (IMPORTANT)

Mode	Meaning

r	Read only

w	Write (overwrite)

a	Append

x	Create new file

r+	Read + Write

w+	Write + Read

a+	Append + Read

b	Binary mode

```python
f = open("myfile.txt", "r")
print(f.read())
f.close()
```

```python
with open("myfile.txt","a") as file:
    file.write("\n good afternoon sir how are ")
```

```python
with open("reddy.txt","a") as f:
    f.write("VARIABLE ARE CONTAINER THAT HOLDS THE DATA VALUES\n FUNCTION ARE BLOCK OF CODE THAT PERFORM A SPECIFIC TASK")
```

```python
with open("reddy.txt","r") as file:
    read=file.read()
    print(read)
```

```python
try:
    with open("newfile.txt","x") as f:
        f.write("HEY MY FRIND HOW ARE YOU AND HOW WAS YOUR STUDIES GOING ON \n BEST OF LUCK FOR YOUR EXAMS ")
        print("FILE CREATED SUCCESFULY")
except FileExistsError:
    print("FILE ALREADY EXISTS")
```

```python
with open("newfile.txt","r") as f:
    print(type(f))
    f.seek(6)
    print(f.tell())
    line=f.readline()
    print(line)
```
