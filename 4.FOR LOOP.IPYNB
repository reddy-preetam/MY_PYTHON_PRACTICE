**A for loop is used to repeat work again and again.

It runs the code a fixed number of times.

It takes values one by one from a list or range.

It saves time by avoiding repeated code.

It makes programs simple and easy to understand.**

```python
fruits=["Apple","Banana","Gava","Kiwi"]
for items in fruits:
    print(fruits)
```

```python
for i in range(1,10):
    print(i)
```

```python
number=[1,2,3,4,5,6,7,8]
for numbers in number:
    if numbers==6:
        break
    if numbers % 2==0:
        continue
    print(number)
```

```python
fruits=["Apple","Banana","Bava","Kiwi"]
for index,fruits in enumerate(fruits):
    print(f"INDEX :- {index} \t FRUITS :- {fruits}")
```

```python
rows=4
columns=2
for i in range(rows):
    for j in range(columns):
        print(f" ROWS ARE {rows} \t COLUMNS ARE {columns}")
```

```python
fruits=["Apple","Banana","Gava","Kiwi"]
colour=["RED","YELLOW","GREEN","LIGHT GREEN"]
for fruits,colour in zip (fruits,colour):
    print(f" FRUIT :- {fruits} its COLOUR {colour}")
```

```python
num=[1,2,3,4,5,6,7]
for i in range (len(num)):
    num[i]=num[i]**2
    print(num)
```

```python
person={"name":"preetam","age":23,"city":"bidar"}
for key in person:
    print(key)
```

```python
person={"name":"preetam","age":23,"city":"bidar"}

for value in person.values():
    print(value)
```

```python
person={"name":"preetam","age":23,"city":"bidar"}

for key,value in person.items():
    print(f"{key} {value}")
```
