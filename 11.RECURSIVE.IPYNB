```python
def factorial(n):
    if (n==0 or n==1): #BASE CASE
        return 1
    else:
        return n*factorial(n-1) # SAMLLER SUB PROGTRAM / ITS RECURSIVE CASE
print(factorial(5))
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```

```python
def reverse_string(s):
    if len(s)==0:
        return s
    else:
        return s[-1]+ reverse_string (s[:-1])
print(reverse_string("PYTHON"))
```
