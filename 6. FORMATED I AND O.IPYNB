**Formatted input means taking input from the user in a specific format.

It allows reading multiple values at once.

Helps convert input into required data types (int, float, string).

In Python, input() is used for formatted input.

Example: a, b = map(int, input().split())

It makes programs cleaner and shorter.

Used when input has spaces or patterns.

Reduces chances of wrong data type errors.

Very useful in competitive programming.

Makes input handling efficient and readable**
