**"""Slicing is used to get a part of a sequence (string, list, tuple).

It uses the syntax [start : stop : step].

Start index is included, stop index is excluded.

Slicing does not change the original data.

It is commonly used for data extraction and manipulation"""**

```python
str="reddy preetam"
print(str[1:4])
```

```python
str="preetam reddy "
print(str[0:3])
print(str[4:7])
print(str[8:11])
print(str[12:14])
```

```python
str="im studying in collge"
print(str.endswith("lge"))
```

```python
nums = [10, 20, 30, 40, 50, 60]

print(nums[1:4])    # [20, 30, 40]  (index 1 se 3 tak)
print(nums[:3])     # [10, 20, 30]  (0 se 2 tak)
print(nums[2:])     # [30, 40, 50, 60]  (index 2 se end tak)
print(nums[::2])    # [10, 30, 50]  (step = 2)
print(nums[::-1])   # [60, 50, 40, 30, 20, 10]  (reverse list)
```

```python
my_tuple = (1, 2, 3, 4, 5, 6)

print(my_tuple[0:3])   # (1, 2, 3)
print(my_tuple[-3:])   # (4, 5, 6)  (last 3 elements)
print(my_tuple[1:5:2]) # (2, 4)  (skip 1 element in between)
```

```python
text = "DataScience"

print(text[0:4])  
print(text[:6])     
print(text[4:])    
print(text[:6])   
print(text[::-1])   # (reverse string)    
```
