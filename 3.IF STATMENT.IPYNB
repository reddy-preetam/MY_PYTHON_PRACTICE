**The else statement is used with an if statement.

It runs when the if condition is False.

The else block provides an alternative path of execution.

It helps handle cases where the if condition is not satisfied.

The else statement makes decision-making in programs complete and clear.**

```python
# List of marks
marks = [77, 97, 64, 85, 55]

# Function to assign grades
def grade(m):
    if m >= 90:
        return 'A'
    elif 80 <= m < 90:
        return 'B'
    elif 70 <= m < 80:
        return 'C'
    elif 60 <= m < 70:
        return 'D'
    else:
        return 'F'

# Use map to apply grade function to all marks
grades = list(map(grade, marks))  # map object ko list me convert karna hota hai

# Print results
print("Exam scores:", marks)
print("Grades:", grades)
```
