What is a string?
A string is a sequence of characters used to store text.

Written inside quotes
Strings are written using single quotes ' ' or double quotes " ".

Example

name = "Reddy"


Strings are immutable


Once a string is created, its characters cannot be changed.

Indexing is supported
Each character has an index starting from 0.

name[0]   # R


Slicing is possible
You can get a part of a string.

name[0:3]   # Red


Length of string
Use len() to find the length of a string.

len(name)


String concatenation
Strings can be joined using the + operator.

"Hello" + " World"


String methods
Python provides many string methods like upper(), lower(), strip(), replace().

Looping through a string
You can loop through each character in a string.

for ch in name:
    print(ch)
