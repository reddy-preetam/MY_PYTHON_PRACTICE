```python
# 🔹 1. Positional Arguments

# Values are passed in the same order as parameters.

# Order is important
```

```python
def add (a,b):
    return a+b
result=add(1,2) 
print(result)
```

```python
# 🔹 2. Keyword Arguments

# Arguments are passed using parameter names.

# Order is not importa
```

```python
def mul(a,b):
    return a*b
result=mul(b=12,a=22)
print(result)
```

```python
# 🔹 3. Default Arguments

# Parameters have default values.

# If value is not passed, default is used
```

```python
def add(a,b=22):
    return a+b
result=add(2)
print(result)
```

```python
# 🔹 4. Variable-length Arguments (*args)

# Used when number of arguments is unknown.

# Stored as a tuple
```

```python
def fun(normal,*args):
    print(normal)
    for i in args:
        print(i)
names=["PREETAM REDDY","SUBHASH"]
normal="THEY BOTH ARE GOOD FRIENDS :"
fun(normal,*names)
```

```python
# 🔹 5. Keyword Variable-length Arguments (**kwargs)

# Accepts multiple keyword arguments.

# Stored as a dictionary
```

```python
def fun(normal,*args,**kwargs):
    print(normal)
    
    for item in args:
        print(item)

    for key,value in kwargs.items():
        print(f"{key}:{value}")

    names=["PREETAM REDDY","VIRAT KOLI"]
    kw = {"PREETAM": "FAN", "FAV_BATTER":"VIRAT KHOLI"}
    normal = "MY FAV BATTER IN INDIAN TEAM:"
    fun(normal,*names,**kw)
```

```python
def fun(normal, *args, **kwargs):
    print(normal)

    for item in args:
        print(item)

    for key, value in kwargs.items():
        print(f"{key} : {value}")


names = ["PREETAM REDDY", "VIRAT KHOLI"]
kw = {"PREETAM": "FAN", "FAV_BATTER":"VIRAT KHOLI"}
normal = "MY FAV BATTER IN INDIAN TEAM:"

fun(normal, *names, **kw)
```
