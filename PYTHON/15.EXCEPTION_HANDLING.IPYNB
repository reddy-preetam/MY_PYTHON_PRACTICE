What is Exception?

An exception is an error that occurs while program is running, and it stops normal execution

Common Built-in Exceptions 

Exception	Meaning

ZeroDivisionError	0 se divide

ValueError	Wrong value

TypeError	Wrong data type

IndexError	List index galat

KeyError	Dictionary key missing

NameError	Variable not defined

FileNotFoundError	File nahi mili


TRY BLOCK – 

The try block contains code that may cause an error.

It is used to test risky or unsafe code.

The program stops executing the try block if an exception occurs.

A try block must be followed by at least one except block.

It helps prevent the program from crashing.

EXCEPT BLOCK – 

The except block handles errors raised in the try block.

It runs only when an exception occurs.

It prevents the program from terminating abnormally.

It can handle specific or all types of exceptions.

Multiple except blocks can be used with one try block

```python
num1=int(input("enetr a number"))
num2=int(input("enetr a number"))
try:
    num1/num2

except ZeroDivisionError:
    print("YOU CANT DIVIDE BY ZERO :-",num1,num2)
```

```python
try:
    x = int(input("Enter a number: "))
    print(10 / 5)
except ZeroDivisionError:
    print("You cannot divide by zero!")
except ValueError:
    print("PLEASE ENETR A VALID NUMBER :")
```

```python
# Input
try:
    f = open("data.txt")   # may raise FileNotFoundError
except FileNotFoundError:
    print("File missing.")
else:
    print("File opened successfully.")
    f.close()
 
finally:
    print("FINALLY")
```

```python
""" What is Nested tryexcept?

A nested tryexcept means a try block inside another try block.
It is used when different parts of code need separate error handling."""
```

```python
try:
    a = int(input("Enter number: "))
    try:
        b = 10 / a
        print("Result:", b)
    except ZeroDivisionError:
        print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

```python
try:
    lst = [10, 20, 30]
    try:
        print(lst[5])
    except IndexError:
        print("Index out of range")
except Exception:
    print("Outer exception handled")
```

```python
""" What is Exception Chaining?

Exception chaining means linking one exception to another exception.

👉 When one error occurs, it causes another error, and Python keeps both errors connected.

🔹 Why Exception Chaining is Used?

To know the real root cause of an error

To give clear debugging information

To show original error + new error together

 How Exception Chaining Works?

Python uses the keyword from to chain exceptions."""
```

```python
try:
    x = int("abc")
except ValueError as e:
    raise TypeError("Conversion failed") from e
```
