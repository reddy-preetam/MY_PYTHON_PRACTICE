**1_________**

**A function is a block of reusable code that performs a specific task.

Functions help reduce code repetition.

In Python, a function is created using the def keyword.

A function can take input values called parameters.

Functions can return a value using the return statement.

If no value is returned, the function returns None by default.

Functions make programs easy to read and maintain.

A function can be called multiple times in a program.

Python supports built-in functions and user-defined functions.

Functions help break large programs into smaller, manageable parts**

**4_________**

**🔹 Function Call

A function call means executing a function.

It is done by writing the function name with parentheses.

Values can be passed inside the parentheses.

The program jumps to the function definition when called.

After execution, control returns to the calling statement.**

```python
def add(a,b):
    return a+b
add(22,44) #IT IS THE FUNCTION CALL
```

**5_____**

**🔹 Arguments

Arguments are the values passed to a function.

They provide input to the function.

Arguments are written inside parentheses during function call.

A function can have one or more arguments.

Arguments match with parameters in the function definitio**

```python
def result(a,b):
    return a+b
result(1,3)
```

```python
def my(name):
    print(f"my name is {name}")
my("preetam") # IT IS THE FUNCTION AGRUMENT
```

**6_________**

**🔹 Return

return is used to send a value back from a function.

It ends the function execution.

A function can return one or more values.

If no return is written, the function returns None.

Returned value can be stored in a variable**

```python
num=[1,2,3,4,5,6,7,8]
def even(x):
    return x%2==0 # IT IS THE RETURN 
even=list(filter(even,num))
print(f"even {even},num are {num}")
```

```python
city=["bidar","banglore","jaipur","hydrabad","uttarprades","im"]
sort=sorted(city,key=lambda x:len(x))#reverse==true
print(" ACCORDING LENGHT OF CITIES ",sort)
```

```python
marks=[99,88,77,66,55]
def failling(score):
    return score<88
result=filter(failling,marks)
print("failling score",list(result))
```
