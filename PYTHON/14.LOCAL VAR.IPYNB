```python
# A local variable is declared inside a function.

# It can be accessed only within that function.

# Local variables have local scope.

# They are created when the function is called.

# They are destroyed after the function ends.

# Local variables cannot be accessed outside the function.

# A local variable has higher priority than a global variable with the same name.

# Local variables help make code safe and modular.

# They reduce chances of unexpected value changes.

# Local variables can shadow global variables
```

```python
def show():
    x = 10
    print(x)

show()
```

```python
def f1():
    x = 10
    print(x)

def f2():
    x = 20
    print(x)

f1()
f2()
```

```python
x = 100

def test():
    y = 50
    print(y)

test()
print(x)
```

```python
x = 100   # global

def demo():
    x = 50   # local shadows global
    print(x)

demo()
print(x)
```

```python
#🔹 Example 9: Shadow Removed Using global
x = 100

def demo():
    global x
    x = 50
    print(x)

demo()
print(x)
```
