Python provides date and time handling using the datetime module.

The date class is used to work with only date (year, month, day).

The time class is used to work with only time (hour, minute, second).

The datetime class combines both date and time together.

The now() method gives the current date and time.

The today() method gives the current date.

The strftime() method is used to format date and time into readable strings.

The strptime() method converts a string into date/time object.

The timedelta class is used to add or subtract days, hours, minutes.

Date and time are widely used in logging, attendance, billing, and scheduling.

```python
import datetime
```

```python
current_datetime=datetime.datetime.now()
print(current_datetime.year)
print(current_datetime.month)
print(current_datetime.day)
print(current_datetime.hour)
print(current_datetime.minute)
print(current_datetime.second)
print(current_datetime.microsecond)
```

```python
current_datetim=datetime.datetime.now()
formated_datetime=current_datetim.strftime("%Y-%m-%d-%H-%M-%S")
print(formated_datetime)
```

```python
current_datetim=datetime.datetime.now()
new_datetime=current_datetim+datetime.timedelta(days=8)
print(new_datetime)
```

```python
import pytz
```

```python
current_datetim=datetime.datetime.now(pytz.timezone("AMERICA/NEW_YORK"))
print(current_datetim)
```

```python
start_time=datetime.datetime(2026,2,2,)
end_time=datetime.datetime(2026,7,1)
durition=end_time-start_time
print(durition)
```
