What is indexing?
Indexing means accessing individual characters of a string using position numbers.

Index starts from 0
In Python, indexing always starts from 0.

Positive indexing
Left to right indexing is called positive indexing.

text = "Python"
print(text[0])   # P
print(text[3])   # h


Negative indexing
Right to left indexing is called negative indexing.

print(text[-1])  # n
print(text[-2])  # o


Indexing works on strings
Each character in a string has its own index.

Space also has an index
Spaces in a string are also counted as characters.

Immutable behavior
You cannot change a character using indexing.

text[0] = "J"   # Error


IndexError
Using an invalid index gives an error.

print(text[10])  # IndexError


Indexing is read-only
You can only read characters, not modify them.

Used in loops and conditions
Indexing is often used with loops and conditions.


```python
mess="hello world hey"
print(mess[4])
```

```python
mess="hello world hey"
print(mess[-2])
```

```python
mess="hello world hey"
# print(mess[0:8:1])
print(mess[-1:-8:2])
```

```python
mess="hello world"
new_mess=mess[:6]+"preetam"
print(new_mess)
```
