**"""An operator is a symbol or keyword that performs an operation on one or more operands.

Operators are used to manipulate data and variables in a program.

They help in performing arithmetic calculations like addition and subtraction.

Operators are also used to compare values and make decisions.

Common operators include arithmetic, relational, logical, and assignment operators"""**

**2_________**

**"""Arithmetic Operators
Used to perform mathematical calculations such as addition, subtraction, multiplication, and division.

Logical Operators
Used to combine conditions or reverse a condition and return True or False.

Relational (Comparison) Operators
Used to compare two values and check relationships like greater than, less than, or equal to.

Assignment Operators (A =A+B)
Used to assign values to variables or update their existing values.

Bitwise Operators (&),(OR),(| XOR), (NOT ),(LEFT SHIFT <<),(RIGHT SHIFT >>)
Used to perform operations on binary (0 and 1) representations of numbers.

Membership Operators (IN  OR NOT IN)
Used to check whether a value exists in a sequence like a list, tuple, or string.

Identity Operators (IS OR IS NOT)
Used to check whether two variables refer to the same object in memory"""**
