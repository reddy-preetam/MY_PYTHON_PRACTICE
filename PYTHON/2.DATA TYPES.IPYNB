Data Types – 10 Points

Data types define the type of data a variable can store.

They help the computer understand the value of a variable.

Different data types store different kinds of data.

Integer (int) stores whole numbers (e.g., 10, 25).

Float stores decimal numbers (e.g., 3.5, 7.25).

String (str) stores text or characters (e.g., "Hello").

Boolean (bool) stores only True or False.

Data types help in memory management.

They make programs error-free and efficient.

Using correct data types improves program performance.
