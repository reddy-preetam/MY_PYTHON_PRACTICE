
🔹 Generator Function – 10 Points

A generator function is a special function that returns values one by one.

It uses the yield keyword instead of return.

When yield is used, the function pauses its execution.

The function resumes from the same point on the next call.

Generator functions return a generator object.

Values are generated only when needed (lazy execution).

Generators are memory efficient compared to lists.

They are mainly used for large data or infinite sequences.

A generator can be iterated using a for loop.

Once a generator is exhausted, it cannot be reused.

🔹 Generator Function 
A generator function is a special function that returns values one by one.

It uses the yield keyword instead of return.

When yield is used, the function pauses its execution.

The function resumes from the same point on the next call.

Generator functions return a generator object.

Values are generated only when needed (lazy execution).

Generators are memory efficient compared to lists.

A generator can be iterated using a for loop.

Once a generator is exhausted, it cannot be reused.

```python
def number():
    yield 10
    yield 20
    yield 23
    yield 299
R=number()
print(next(R))
print(next(R))
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    yield 4
for value in my_generator():
    print(value)
```
