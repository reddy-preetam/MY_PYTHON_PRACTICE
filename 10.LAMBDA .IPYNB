A lambda function is a small anonymous function (no name).

It is created using the lambda keyword.

Lambda functions can take any number of arguments.

They can have only one expression.

The expression result is automatically returned.

Lambda functions do not use the return keyword.

They are written in one single line.

Used for short, simple operations.

Commonly used with map(), filter(), and reduce().

Lambda functions make code short and concise.

```python
city=["bidar","banglore","rr","bhalki road cross"]
def lenght (city):
    return len (city)>5
result=list(filter(lenght,city))
print(result)
```

```python
city=["bidar","banglore","rr","bhalki road cross"]
def lenght (city):
    return len (city)<10
result=list(filter(lenght,city))
print(result)
```

```python
city=["bidar","banglore","rr","bhalki road cross"]
result=list(filter(lambda city:len(city)>5,city))
print(result)
```

```python
num=[1,2,3,4,5,6,7,8]
even=list(filter(lambda x:x%2==0,num))
print(even)
```

reduce() is used to apply a function repeatedly to elements of a sequence.

It reduces multiple values to a single value.

It is available in the functools module.

It takes two arguments at a time.

The result of each step is passed to the next step.

It works on sequences like list, tuple, set.

Commonly used for sum, product, max, min operations.

A lambda function is often used with reduce().

It returns one final value.

It is useful when a cumulative result is needed

```python
from functools import reduce
num=[1,2,3,4,5,6,7,8]
result=reduce(lambda x,y:x*y,num)
print(result)
```

🔹 map() Function – 10 Points

map() is a built-in function in Python.

It is used to apply a function to each element of an iterable.

An iterable can be a list, tuple, set, or string.

Syntax of map() is:

map(function, iterable) IT MEANS EK EK KARKE VALUE DENA


The function defines what operation will be performed.

map() returns a map object, not a list directly.

To see the result, we usually convert it into a list using list().

map() helps write shorter and cleaner code than loops.

It is commonly used with lambda (anonymous) functions.

map() can work with multiple iterables at the same time

```python
num=[1,2,3,4,5,6,7,8]
def squer(x):
    return x*x
result=list(map(squer,num))
print(result)
```

```python
num=[1,2,3,4,5,6,7,8]
even=list(map(lambda x:x*x,num))
print(even)
```
